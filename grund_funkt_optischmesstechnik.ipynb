```python
import numpy as np

data = [12,23,45,67]

print("Das ist es Median: " , np.median(data))
print("Dast ist Mittelwert: ", np.mean(data))
print("Das ist Standardabweichung: " , np.std(data, ddof=1))
```

```python
# Normalverteilung

import numpy as numpy 
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameter 
mu = 0 # Mittelwert
sigma = 1 # Standardabweichung

# x-Werte
x = np.linspace(-5,5,1000)

# Dichtefunktion berechnen
y = norm.pdf(x,mu,sigma)

fig,ax = plt.subplots(2,1)
fig.tight_layout(pad = 3.0) # add spacing between graphs
ax = ax.flatten()
plt.sca(ax[0])
plt.plot(x,y,label=f"mu_={mu}, sigma_ = {sigma}")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()

# 1000 Zufallswerte aud N 
daten = np.random.normal(mu,sigma,1000)
plt.sca(ax[1])
plt.hist(daten,bins=30,density= True,alpha = 0.6, color='g')
plt.show()
```
