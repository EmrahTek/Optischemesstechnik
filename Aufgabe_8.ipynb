```python
import numpy as np
#Beispielwerte
Delta_a = 0.2 # Fehler von an
Delta_b = 0.1 # Fehler von b
Delta_c = 0.3 # Fehler von c

# Fehlerfortpflanzung
Delta_y = np.sqrt(Delta_a**2+ Delta_b**2 + Delta_c**2)

print(f"Gesamtfehler delta y = {Delta_y:.3f}")
```
